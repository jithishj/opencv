```python
import cv2
import numpy as np
def roi(img, pts):
    mask=np.zeros_like(img)
    cv2.fillPoly(mask, [pts], (255,255,255))
    mask_img=cv2.bitwise_and(img, mask)
    return mask_img
def draw_line(img, lines):
    img=np.copy(img)
    blank_image=np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    for (x1, y1, x2, y2) in lines[0]:
        cv2.line(blank_image, (x1, y1), (x2, y2), (0, 255, 0), thickness=3)
    img=cv2.addWeighted(img, .8, blank_image, 1, 0.0)
    return img
```

```python
def process(img):
    print(img.shape)
    height=img.shape[0]
    width=img.shape[1]
    pts=[(0, height), (width/2, height/2), (width, height)]
    gray_img=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    canny_img=cv2.Canny(gray_img, 100, 200)
    crop_img=roi(canny_img, np.array(pts, np.int32))
    lines=cv2.HoughLinesP(crop_img, 3, np.pi/60, 200, minLineLength=50, maxLineGap=30)
    image_with_lines=draw_line(img, lines)
    return image_with_lines

    
```

```python
cap=cv2.VideoCapture(r'C:/Users/jithi/Desktop/test_video.mp4')
while (cap.isOpened()):
    ret, frame=cap.read()
    frame=process(frame)
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) ==27:
        break
cap.release()
cv2.destroyAllWindows()

    
```

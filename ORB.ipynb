```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
orb=cv2.ORB_create()
```

```python
query_img=cv2.imread(r'C:/Users/jithi/Desktop/nasa_logo.png')
train_img=cv2.imread(r'C:/Users/jithi/Desktop/nasa.jpg')
```

```python
kp0, des0=orb.detectAndCompute(query_img, None)
kp1, des1=orb.detectAndCompute(train_img, None)
```

```python
matcher=cv2.BFMatcher()
matches=matcher.match(des0, des1)
```

```python
final_img=cv2.drawMatches(query_img, kp0, train_img, kp1, matches[:20], None)
```

```python
plt.imshow(final_img)
```

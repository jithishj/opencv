```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread(r'C:/Users/jithi/Desktop/seven.png')
gray=cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
kernel=np.ones((5,5), dtype='uint8')
```

```python
dilated=cv2.dilate(image, kernel, 2)
```

```python
plt.imshow(dilated, cmap='gray')
```

```python
eroded=cv2.erode(dilated, kernel, 5)
```

```python
plt.imshow(eroded)
    
```

```python
opening=cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening)
```

```python
closing=cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing)
```
